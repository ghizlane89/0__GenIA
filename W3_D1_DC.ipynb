```python
# Étape 1 : Génération des données

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Fixer la graine pour des résultats reproductibles
np.random.seed(0)

# Noms des villes et des mois
cities = ['Paris', 'New York', 'Tokyo', 'Cairo', 'Moscow',
          'Sydney', 'Rio', 'Toronto', 'Beijing', 'Cape Town']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Générer des températures entre -5 et 35°C
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

# Créer le DataFrame
df = pd.DataFrame(temperature_data, index=cities, columns=months)

print (df)
```

```python
#Étape 2 : Analyse des données

# Moyenne annuelle par ville
annual_avg = df.mean(axis=1)

# Ville la plus chaude
hottest_city = annual_avg.idxmax()

# Ville la plus froide
coldest_city = annual_avg.idxmin()

# Affichage des résultats
print(f" Ville la plus chaude : {hottest_city} ({annual_avg[hottest_city]:.2f} °C)")
print(f" Ville la plus froide : {coldest_city} ({annual_avg[coldest_city]:.2f} °C)")
```

```python
#Étape 3 : Visualisations

# Courbe pour 4 villes

plt.figure(figsize=(12, 6))
df.loc[['Paris', 'Tokyo', 'Cairo', 'Moscow']].T.plot(marker='o')
plt.title('Évolution des températures mensuelles (4 villes)')
plt.ylabel('Température moyenne (°C)')
plt.xlabel('Mois')
plt.grid(True)
plt.legend(title='Villes')
plt.tight_layout()
plt.show()

```

```python
# Moyenne annuelle par ville

plt.figure(figsize=(10, 5))
annual_avg.sort_values().plot(kind='bar', color='skyblue')
plt.title('Température annuelle moyenne par ville')
plt.ylabel('Température moyenne (°C)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
summary = {
    'Ville la plus chaude': hottest_city,
    'Température moyenne': round(annual_avg[hottest_city], 2),
    'Ville la plus froide': coldest_city,
    'Température moyenne': round(annual_avg[coldest_city], 2)
}

summary
```
