```python
#Exercice 1:


# Tache 1:
# Importance de la visualisation de données :
# Permet de voir rapidement les tendances et les relations entre variables
# Aide à détecter les anomalies ou erreurs dans les données
# Facilite la compréhension de données complexes
# Améliore la prise de décision grâce à une vue claire
# Utile pour communiquer les résultats à un public non technique

# Tache 2:

# Un graphique en ligne (ou graphique linéaire) est utilisé pour montrer
# l’évolution d’une variable dans le temps. Il relie des points de données
# successifs avec des lignes pour illustrer des tendances, des hausses ou
# des baisses. C’est un outil idéal pour analyser des séries temporelles
# comme la variation des températures, des ventes, ou du trafic web au fil
# des jours, mois ou années.





```

```python
#Exercice 2:


import matplotlib.pyplot as plt

x = ['lundi', 'mardi', 'mercredi', 'jeudi','vendredi', 'samedi', 'dimanche' ]
y = [72, 74, 76, 80, 82, 78, 75]
plt.plot(x, y)


plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o')
plt.title("Évolution de la temperature en fonction du jour de la semaine")
plt.xlabel("Jour de la semaine")
plt.ylabel("Temperature (°F)")

plt.show()
```

```python
#Exercice 5


from google.colab import drive
drive.mount('/content/drive')


import pandas as pd

file_path = '/content/drive/My Drive/Datasets/Student Mental health.csv'
df = pd.read_csv(file_path)
df.head()


import seaborn as sns
import matplotlib.pyplot as plt

print(df.columns)
df.head()




```

```python
# cr
anxious = df[df['Do you have Anxiety?'] == 'Yes']
#print(anxious)



```

```python
group_counts = anxious.groupby('Choose your gender').size()
#print(group_counts)

proportions = group_counts / group_counts.sum()
#print(proportions)

counts = proportions.reset_index()
#print(counts)


counts.columns = ['Gender', 'Proportion']

sns.barplot(data=counts, x='Gender', y='Proportion', palette='Set2')
plt.title("Proportion d'étudiants souffrant d'anxiété selon le genre")
plt.xlabel("Genre")
plt.ylabel("Proportion (%)")
plt.ylim(0, 1)
plt.show()


```
