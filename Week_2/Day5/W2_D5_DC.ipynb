```python
#Convertir les colonnes de dates en datetime

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.head()
```

```python
#Calcul du Profit Margin en pourcentage
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100

#Extraction de l‚Äôann√©e et du mois de la commande
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df.head()
```

```python
# Tendance des ventes mensuelles

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Agr√©ger par mois
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum()

# Tracer
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Tendance globale des ventes mensuelles')
plt.xlabel('Mois')
plt.ylabel('Ventes totales')
plt.grid(True)

# Ajouter un filtre interactif pour la cat√©gorie
import ipywidgets as widgets
from IPython.display import display

categories = df['Category'].unique()
category_dropdown = widgets.Dropdown(options=categories, description='Cat√©gorie:')

def plot_trend(cat):
    filtered_df = df[df['Category'] == cat]
    monthly = filtered_df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum()
    plt.figure(figsize=(12,6))
    plt.plot(monthly.index, monthly.values, marker='o')
    plt.title(f'Tendance des ventes mensuelles pour {cat}')
    plt.xlabel('Mois')
    plt.ylabel('Ventes totales')
    plt.grid(True)
    plt.show()

widgets.interact(plot_trend, cat=category_dropdown)
```

```python
#Vente par √©tat (top N)
import seaborn as sns

# Agr√©ger par √©tat
sales_by_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Slider pour choisir top N
top_n_slider = widgets.IntSlider(min=1, max=50, step=1, value=10, description='Top N States:')

def plot_top_states(n):
    top_states = sales_by_state.head(n)
    plt.figure(figsize=(12,6))
    sns.barplot(x=top_states.values, y=top_states.index)
    plt.title(f'Top {n} √âtats par ventes')
    plt.xlabel('Ventes')
    plt.ylabel('√âtat')
    plt.show()

widgets.interact(plot_top_states, n=top_n_slider)
```

```python
#Produits les plus profitables
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 produits les plus profitables')
plt.xlabel('Profit total')
plt.ylabel('Produit')
for index, value in enumerate(top_products.values):
    plt.text(value, index, f'{value:.2f}', va='center')
plt.show()
```

```python
# Analyse de l‚Äôimpact du discount
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='Discount',
    y='Profit',
    hue='Category',
    alpha=0.9
)
# Ajout d‚Äôune ligne de tendance globale
sns.regplot(
    data=df,
    x='Discount', y='Profit',
    scatter=False, color='black'
)
plt.title('Relation entre Discount et Profit par Cat√©gorie')
plt.xlabel('Remise (%)')
plt.ylabel('Profit')
plt.show()
```

La pr√©sence d'object dans la colonne Dtype sugg√®re de proc√©der √† de l'encoding

```python
## A] Encodage Label (pour des variables avec peu de cat√©gories, par exemple 'Region', 'Segment', 'Ship Mode')
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple pour 'Region'
le_region = LabelEncoder()
df['Region_encoded'] = le_region.fit_transform(df['Region'])

# Pour 'Segment'
le_segment = LabelEncoder()
df['Segment_encoded'] = le_segment.fit_transform(df['Segment'])

# Pour 'Ship Mode'
le_ship_mode = LabelEncoder()
df['Ship Mode_encoded'] = le_ship_mode.fit_transform(df['Ship Mode'])

## B] Encodage One-Hot (pour variables avec plusieurs cat√©gories, par exemple 'Category', 'Sub-Category')
# Encodage One-Hot pour 'Category' et 'Sub-Category'
category_dummies = pd.get_dummies(df['Category'], prefix='Category')
sub_category_dummies = pd.get_dummies(df['Sub-Category'], prefix='SubCategory')

df.info()
```

```python
# Ajoute ces colonnes au DataFrame principal
df = pd.concat([df, category_dummies, sub_category_dummies], axis=1)

# Affichage graphique
sns.boxplot(x='Region_encoded', y='Sales', data=df)
plt.xticks(ticks=range(len(le_region.classes_)), labels=le_region.classes_)
plt.title('Distribution des ventes par r√©gion')
plt.show()
```
