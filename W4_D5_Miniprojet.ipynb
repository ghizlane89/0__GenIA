```python
# InterprÃ©tations et choix dâ€™imputation :
Pour les variables numÃ©riques comme trestbps, chol, oldpeak et ca, une imputation par la moyenne ou la mÃ©diane est envisageable. NÃ©anmoins, pour certaines variables spÃ©cifiques comme ca (qui indique le nombre de vaisseaux colorÃ©s), il peut Ãªtre pertinent dâ€™Ã©tudier leur distribution avant dâ€™imputer, ou dâ€™envisager une stratÃ©gie dâ€™imputation plus adaptÃ©e (ex. : mode, valeur dominante ou imputation par classe).

La variable id, probablement utilisÃ©e comme identifiant unique, nâ€™apporte pas dâ€™information utile pour la modÃ©lisation. Elle est donc supprimÃ©e.

La variable num, qui reprÃ©sente la cible (prÃ©sence ou absence de maladie cardiaque), nâ€™est jamais imputÃ©e. Elle est conservÃ©e telle quelle pour ne pas introduire de biais.

# Imputation conditionnelle :
Pour certaines variables, comme chol, on peut supposer que les valeurs manquantes dÃ©pendent de lâ€™Ã©tat de santÃ© du patient (malade ou non). Dans ce cas, il est judicieux dâ€™utiliser une stratÃ©gie dâ€™imputation conditionnelle : on remplit les valeurs manquantes par la moyenne ou la mÃ©diane calculÃ©e dans chaque groupe de la variable cible. Cela permet de respecter les relations sous-jacentes entre les variables et dâ€™Ã©viter un lissage artificiel des diffÃ©rences entre classes.

# Recommandation gÃ©nÃ©rale :
Lorsquâ€™une variable est utilisÃ©e comme cible de prÃ©diction (ici, num), il est essentiel de ne jamais l'utiliser pour imputer ses propres valeurs. En revanche, il est tout Ã  fait pertinent de sâ€™appuyer sur num pour guider lâ€™imputation d'autres variables, ce qui peut amÃ©liorer la cohÃ©rence des donnÃ©es et la performance des modÃ¨les en aval.

```

```python
import pandas as pd

# ğŸ”¹ Suppression de la colonne 'id' (identifiant inutile pour la modÃ©lisation)
df.drop('id', axis=1, inplace=True)

# ğŸ”¹ Imputation des valeurs manquantes pour les variables numÃ©riques

# â¤ 'trestbps' : tension artÃ©rielle au repos â€” imputation par la mÃ©diane globale
df['trestbps'].fillna(df['trestbps'].median(), inplace=True)

# â¤ 'chol' : taux de cholestÃ©rol â€” imputation conditionnelle par mÃ©diane selon la classe cible 'num'
for target_value in df['num'].unique():
    mask = df['num'] == target_value
    median_chol = df.loc[mask, 'chol'].median()
    df.loc[mask & df['chol'].isnull(), 'chol'] = median_chol

# â¤ 'thalch' : frÃ©quence cardiaque maximale â€” imputation par la mÃ©diane globale
df['thalch'].fillna(df['thalch'].median(), inplace=True)

# â¤ 'oldpeak' : mesure de dÃ©pression ST â€” imputation par la mÃ©diane
df['oldpeak'].fillna(df['oldpeak'].median(), inplace=True)

# â¤ 'ca' : nombre de vaisseaux principaux visibles aprÃ¨s colorant â€” imputation par la valeur la plus frÃ©quente (mode)
df['ca'].fillna(df['ca'].mode()[0], inplace=True)

# ğŸ” VÃ©rification finale : nombre de valeurs manquantes par colonne aprÃ¨s traitement
print("Nombre de NaN restant aprÃ¨s imputation :")
print(df.isnull().sum())
```

## ModÃ¨l Training

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.impute import SimpleImputer

# ğŸ“Œ Ã‰tape 1 : VÃ©rification des valeurs manquantes dans les features
print("ğŸ” Nombre de NaN avant imputation :")
print(X.isnull().sum())

# ğŸ“Œ Ã‰tape 2 : Imputation des valeurs manquantes avec la mÃ©diane
imputer = SimpleImputer(strategy='median')

# Application de l'imputation Ã  toutes les colonnes de X
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# ğŸ“Œ Ã‰tape 3 : VÃ©rification aprÃ¨s imputation
print("\nâœ… Nombre de NaN aprÃ¨s imputation :")
print(X_imputed.isnull().sum())

# âš ï¸ Si des NaN subsistent ici, il peut sâ€™agir dâ€™un problÃ¨me liÃ© Ã  des colonnes non numÃ©riques ou Ã  la cible

# ğŸ“Œ Ã‰tape 4 : SÃ©paration du dataset en ensemble d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y, test_size=0.3, random_state=42, stratify=y)

# ğŸ“Œ Ã‰tape 5 : EntraÃ®nement du modÃ¨le de rÃ©gression logistique
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# ğŸ“Œ Ã‰tape 6 : PrÃ©dictions et Ã©valuation du modÃ¨le
y_pred = model.predict(X_test)

# Affichage des mÃ©triques de performance
print("\nğŸ“Š Rapport de classification :")
print(classification_report(y_test, y_pred))

print("\nğŸ§® Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ğŸ“Œ Ã‰tape 1 : PrÃ©dictions sur lâ€™ensemble de test
y_pred = model.predict(X_test)

# ğŸ“Œ Ã‰tape 2 : Calcul de lâ€™exactitude (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"ğŸ¯ Exactitude (Accuracy) : {accuracy:.2f}")

# ğŸ“Œ Ã‰tape 3 : Calcul des autres mÃ©triques de performance (pondÃ©rÃ©es selon les classes)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')

print(f"ğŸ“Œ PrÃ©cision (Precision) : {precision:.2f}")
print(f"ğŸ“Œ Rappel (Recall) : {recall:.2f}")
print(f"ğŸ“Œ F1-Score : {f1:.2f}")

# ğŸ“Œ Ã‰tape 4 : Affichage de la matrice de confusion
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(cm.shape[1]),
            yticklabels=range(cm.shape[0]))
plt.xlabel('PrÃ©dictions')
plt.ylabel('VÃ©ritables valeurs')
plt.title('ğŸ§® Matrice de confusion')
plt.show()
```

```python
# InterprÃ©tation des rÃ©sultats du modÃ¨le
Exactitude (Accuracy) : 61 %
Le modÃ¨le effectue une prÃ©diction correcte dans 61 % des cas. Cela indique une performance correcte, mais encore perfectible : le modÃ¨le rÃ©ussit Ã  identifier la prÃ©sence ou lâ€™absence de maladie chez un peu plus de la moitiÃ© des patients.

PrÃ©cision : 0.57
Parmi les patients que le modÃ¨le prÃ©dit comme malades, seuls 57 % le sont rÃ©ellement. Cette prÃ©cision modÃ©rÃ©e suggÃ¨re une tendance Ã  gÃ©nÃ©rer des faux positifs (patients identifiÃ©s Ã  tort comme atteints).

Rappel : 0.61
Le modÃ¨le dÃ©tecte environ 61 % des vrais cas de maladie cardiaque. Autrement dit, prÃ¨s de 4 patients malades sur 10 ne sont pas identifiÃ©s (faux nÃ©gatifs), ce qui peut poser problÃ¨me dans un contexte mÃ©dical sensible.

F1-Score : 0.59
Le F1-score, qui combine prÃ©cision et rappel, confirme une performance globale moyenne. Le modÃ¨le parvient Ã  un compromis acceptable entre les erreurs de surdÃ©tection (faux positifs) et de sous-dÃ©tection (faux nÃ©gatifs).

Matrice de confusion :
La matrice rÃ©vÃ¨le des confusions entre certaines classes. Cela peut sâ€™expliquer par la proximitÃ© des caractÃ©ristiques cliniques entre groupes de patients, mais aussi par un Ã©ventuel dÃ©sÃ©quilibre des classes dans le jeu de donnÃ©es. Ce type de confusion est frÃ©quent dans les problÃ¨mes de classification multi-classes en contexte mÃ©dical.
```
