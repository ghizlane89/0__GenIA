```python
# 📦 Importation des bibliothèques indispensables pour l’analyse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 🔍 1. Analyse exploratoire des données (EDA)

# Affichage du nombre de lignes et de colonnes du dataset
print(f"Nombre de lignes : {df.shape[0]}, Nombre de colonnes : {df.shape[1]}")

# ➤ Visualisation de la distribution de l’âge des patients
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution de l\'âge')
plt.xlabel('Âge')
plt.ylabel('Nombre de patients')
plt.show()

# ➤ Répartition de la variable cible : présence ou non d'une maladie cardiaque
plt.figure(figsize=(6, 4))
sns.countplot(x='num', data=df)
plt.title('Répartition de la variable cible (maladie cardiaque)')
plt.xlabel('Présence de maladie (0 = non, 1 = oui)')
plt.ylabel('Nombre de patients')
plt.show()

# ➤ Visualisation des corrélations entre variables numériques
numeric_cols = df.select_dtypes(include='number').columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de corrélation entre les variables numériques')
plt.show()

# 🧹 2. Prétraitement des données

# ➤ Identification des colonnes de type catégoriel
categorical_cols = df.select_dtypes(include='object').columns
print("\nColonnes catégoriques :", categorical_cols)

# ➤ Encodage des variables catégoriques avec LabelEncoder
# (utile si certaines variables sont au format texte)
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# ➤ Aperçu des données après encodage
print("\nDonnées après encodage des variables catégoriques :")
print(df.head())

# ➤ Standardisation des variables numériques (hors identifiants)
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include='float64').columns
numeric_cols = [col for col in numeric_cols if col != 'id']  # exclusion des identifiants si présents

# ➤ Application de la normalisation
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# ➤ Vérification des données après normalisation
print("\nDonnées après normalisation :")
print(df.head())

# 🎯 3. Séparation des variables explicatives (X) et de la variable cible (y)
X = df.drop('num', axis=1)
y = df['num']

# ➤ Le jeu de données est maintenant prêt pour l'entraînement d’un modèle
print("\nFeatures (X) et variable cible (y) préparées pour la modélisation.")
```

```python
import pandas as pd

# Vérifier s'il y a des NaN dans l'ensemble du dataset
nan_presence = df.isnull().any().any()

if nan_presence:
    print("Il y a des valeurs NaN dans le dataset.")
    # Optionnel : afficher le nombre de NaN par colonne
    print("Nombre de NaN par colonne :")
    print(df.isnull().sum())
else:
    print("Il n'y a pas de valeurs NaN dans le dataset.")
```

```python
# Interprétations et choix d’imputation :
Pour les variables numériques comme trestbps, chol, oldpeak et ca, une imputation par la moyenne ou la médiane est envisageable. Néanmoins, pour certaines variables spécifiques comme ca (qui indique le nombre de vaisseaux colorés), il peut être pertinent d’étudier leur distribution avant d’imputer, ou d’envisager une stratégie d’imputation plus adaptée (ex. : mode, valeur dominante ou imputation par classe).

La variable id, probablement utilisée comme identifiant unique, n’apporte pas d’information utile pour la modélisation. Elle est donc supprimée.

La variable num, qui représente la cible (présence ou absence de maladie cardiaque), n’est jamais imputée. Elle est conservée telle quelle pour ne pas introduire de biais.

# Imputation conditionnelle :
Pour certaines variables, comme chol, on peut supposer que les valeurs manquantes dépendent de l’état de santé du patient (malade ou non). Dans ce cas, il est judicieux d’utiliser une stratégie d’imputation conditionnelle : on remplit les valeurs manquantes par la moyenne ou la médiane calculée dans chaque groupe de la variable cible. Cela permet de respecter les relations sous-jacentes entre les variables et d’éviter un lissage artificiel des différences entre classes.

# Recommandation générale :
Lorsqu’une variable est utilisée comme cible de prédiction (ici, num), il est essentiel de ne jamais l'utiliser pour imputer ses propres valeurs. En revanche, il est tout à fait pertinent de s’appuyer sur num pour guider l’imputation d'autres variables, ce qui peut améliorer la cohérence des données et la performance des modèles en aval.

```

```python
import pandas as pd

# 🔹 Suppression de la colonne 'id' (identifiant inutile pour la modélisation)
df.drop('id', axis=1, inplace=True)

# 🔹 Imputation des valeurs manquantes pour les variables numériques

# ➤ 'trestbps' : tension artérielle au repos — imputation par la médiane globale
df['trestbps'].fillna(df['trestbps'].median(), inplace=True)

# ➤ 'chol' : taux de cholestérol — imputation conditionnelle par médiane selon la classe cible 'num'
for target_value in df['num'].unique():
    mask = df['num'] == target_value
    median_chol = df.loc[mask, 'chol'].median()
    df.loc[mask & df['chol'].isnull(), 'chol'] = median_chol

# ➤ 'thalch' : fréquence cardiaque maximale — imputation par la médiane globale
df['thalch'].fillna(df['thalch'].median(), inplace=True)

# ➤ 'oldpeak' : mesure de dépression ST — imputation par la médiane
df['oldpeak'].fillna(df['oldpeak'].median(), inplace=True)

# ➤ 'ca' : nombre de vaisseaux principaux visibles après colorant — imputation par la valeur la plus fréquente (mode)
df['ca'].fillna(df['ca'].mode()[0], inplace=True)

# 🔍 Vérification finale : nombre de valeurs manquantes par colonne après traitement
print("Nombre de NaN restant après imputation :")
print(df.isnull().sum())
```

## Modèl Training

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.impute import SimpleImputer

# 📌 Étape 1 : Vérification des valeurs manquantes dans les features
print("🔍 Nombre de NaN avant imputation :")
print(X.isnull().sum())

# 📌 Étape 2 : Imputation des valeurs manquantes avec la médiane
imputer = SimpleImputer(strategy='median')

# Application de l'imputation à toutes les colonnes de X
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# 📌 Étape 3 : Vérification après imputation
print("\n✅ Nombre de NaN après imputation :")
print(X_imputed.isnull().sum())

# ⚠️ Si des NaN subsistent ici, il peut s’agir d’un problème lié à des colonnes non numériques ou à la cible

# 📌 Étape 4 : Séparation du dataset en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y, test_size=0.3, random_state=42, stratify=y)

# 📌 Étape 5 : Entraînement du modèle de régression logistique
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# 📌 Étape 6 : Prédictions et évaluation du modèle
y_pred = model.predict(X_test)

# Affichage des métriques de performance
print("\n📊 Rapport de classification :")
print(classification_report(y_test, y_pred))

print("\n🧮 Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 📌 Étape 1 : Prédictions sur l’ensemble de test
y_pred = model.predict(X_test)

# 📌 Étape 2 : Calcul de l’exactitude (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"🎯 Exactitude (Accuracy) : {accuracy:.2f}")

# 📌 Étape 3 : Calcul des autres métriques de performance (pondérées selon les classes)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')

print(f"📌 Précision (Precision) : {precision:.2f}")
print(f"📌 Rappel (Recall) : {recall:.2f}")
print(f"📌 F1-Score : {f1:.2f}")

# 📌 Étape 4 : Affichage de la matrice de confusion
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(cm.shape[1]),
            yticklabels=range(cm.shape[0]))
plt.xlabel('Prédictions')
plt.ylabel('Véritables valeurs')
plt.title('🧮 Matrice de confusion')
plt.show()
```

```python
# Interprétation des résultats du modèle
Exactitude (Accuracy) : 61 %
Le modèle effectue une prédiction correcte dans 61 % des cas. Cela indique une performance correcte, mais encore perfectible : le modèle réussit à identifier la présence ou l’absence de maladie chez un peu plus de la moitié des patients.

Précision : 0.57
Parmi les patients que le modèle prédit comme malades, seuls 57 % le sont réellement. Cette précision modérée suggère une tendance à générer des faux positifs (patients identifiés à tort comme atteints).

Rappel : 0.61
Le modèle détecte environ 61 % des vrais cas de maladie cardiaque. Autrement dit, près de 4 patients malades sur 10 ne sont pas identifiés (faux négatifs), ce qui peut poser problème dans un contexte médical sensible.

F1-Score : 0.59
Le F1-score, qui combine précision et rappel, confirme une performance globale moyenne. Le modèle parvient à un compromis acceptable entre les erreurs de surdétection (faux positifs) et de sous-détection (faux négatifs).

Matrice de confusion :
La matrice révèle des confusions entre certaines classes. Cela peut s’expliquer par la proximité des caractéristiques cliniques entre groupes de patients, mais aussi par un éventuel déséquilibre des classes dans le jeu de données. Ce type de confusion est fréquent dans les problèmes de classification multi-classes en contexte médical.
```
