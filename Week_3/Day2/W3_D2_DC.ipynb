```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- 1. Importation et nettoyage des données ---

# Vérifier la présence de valeurs manquantes
print("Résumé des valeurs manquantes par colonne :")
print(df.isnull().sum())

# Gérer les valeurs manquantes
# Par exemple, pour les colonnes numériques, on peut remplir par la moyenne
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
#.fillna() pour remplir les valeurs manquantes avec la moyenne ou le mode ne remplace pas par zéro, mais par la valeur calculée (moyenne ou mode).

# Pour les colonnes catégorielles, on peut remplir par la valeur la plus fréquente
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Vérification après nettoyage
print("Valeurs manquantes après traitement :")
print(df.isnull().sum())

# Conversion en types numériques si nécessaire (par exemple, 'commissioning_year')
df['commissioning_year'] = pd.to_numeric(df['commissioning_year'], errors='coerce')

# --- 2. Analyse exploratoire (EDA) ---

# Résumé statistique des colonnes numériques
print("Statistiques descriptives :")
print(df.describe())

# Distribution du nombre de centrales par pays
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='country')
plt.title("Répartition des centrales par pays")
plt.xticks(rotation=45)
plt.show()

# Distribution par type de combustible principal
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='primary_fuel', order=df['primary_fuel'].value_counts().index)
plt.title("Répartition des centrales par type de combustible")
plt.show()

# --- 3. Analyse statistique ---

# Analyse du rendement par type de combustible
# Supposons que 'generation_gwh_2017' est une mesure de sortie électrique
fuel_types = df['primary_fuel'].unique()

# Calcul des moyennes et écarts-types pour chaque type de fuel
for fuel in fuel_types:
    subset = df[df['primary_fuel'] == fuel]['generation_gwh_2017']
    mean_val = np.mean(subset)
    std_val = np.std(subset)
    print(f"{fuel} : Moyenne = {mean_val:.2f} GWh, Écart-type = {std_val:.2f} GWh")

# Test d'hypothèse : La moyenne de production est-elle différente entre deux types de fuel (exemple : Hydro vs Solar)
hydro = df[df['primary_fuel']=='Hydro']['generation_gwh_2017']
solar = df[df['primary_fuel']=='Solar']['generation_gwh_2017']

# Test t de Student
t_stat, p_value = stats.ttest_ind(hydro, solar, equal_var=False)
print(f"Test t entre Hydro et Solar : t = {t_stat:.3f}, p = {p_value:.3f}")
if p_value < 0.05:
    print("Il existe une différence significative entre les deux groupes.")
else:
    print("Il n'y a pas de différence significative entre les deux groupes.")

# --- 4. Analyse des séries temporelles ---

# Évolution de la capacité installée dans le temps
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='commissioning_year', y='capacity_mw')
plt.title("Évolution de la capacité installée dans le temps")
plt.xlabel("Année de mise en service")
plt.ylabel("Capacité (MW)")
plt.show()

# Analyser la répartition par année de mise en service
plt.figure(figsize=(10,6))
sns.histplot(df['commissioning_year'].dropna(), bins=20)
plt.title("Distribution des années de mise en service")
plt.xlabel("Année")
plt.ylabel("Nombre de centrales")
plt.show()

# --- 5. Visualisations avancées ---

# Distribution géographique des centrales
plt.figure(figsize=(12,8))
plt.scatter(df['longitude'], df['latitude'], c=pd.Categorical(df['primary_fuel']).codes, cmap='tab10', alpha=0.6)
plt.colorbar(ticks=range(len(df['primary_fuel'].unique())), label='Type de fuel')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Répartition géographique des centrales par type de fuel")
plt.show()

# Carte géographique (si basique) avec basemap ou autres outils plus avancés, mais ici simple scatter

# --- 6. Opérations matricielles en contexte réel ---

# Créer une matrice de caractéristiques : capacité, latitude, longitude
features = df[['capacity_mw', 'latitude', 'longitude']].values

# Exemple simple : calcul de la matrice de covariance
cov_matrix = np.cov(features, rowvar=False)
print("Matrice de covariance :")
print(cov_matrix)

# Eigenvalues et eigenvectors pour analyser la variance
eigvals, eigvecs = np.linalg.eig(cov_matrix)
print("Eigenvalues :")
print(eigvals)
print("Eigenvectors :")
print(eigvecs)

# Relevance : eigenvectors indiquent les directions principales de variation dans les données, utiles pour la réduction de dimension ou pour comprendre la distribution spatiale.

# --- 7. Intégration NumPy avec Pandas et Matplotlib ---

# Exemple : filtrage complexe avec NumPy
# Extraire les centrales avec capacité > 1000 MW
mask = np.array(df['capacity_mw']) > 1000
high_capacity_df = df[mask]

# Visualisation sophistiquée : par exemple, histogramme des capacités en utilisant NumPy
capacities = np.array(df['capacity_mw'])
plt.hist(capacities, bins=30, color='orange', alpha=0.7)
plt.title("Distribution des capacités des centrales")
plt.xlabel("Capacité (MW)")
plt.ylabel("Nombre")
plt.show()
```

```python
import pandas as pd

# Sélectionner les colonnes numériques et catégorielles
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Remplacer NaN dans les colonnes numériques par la moyenne
for col in numeric_cols:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)

# Remplacer NaN dans les colonnes catégorielles par le mode
for col in categorical_cols:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

print("Valeurs manquantes après traitement :")
print(df.isnull().sum())
```
