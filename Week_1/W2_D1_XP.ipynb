```python
#  Data type description for the "Global Trends in Mental Health Disorders" dataset

# - index: Quantitative (discrete)
#    Automatically generated row number, used as an identifier.

# - Entity: Qualitative (nominal)
#    Name of the country or region (e.g., Afghanistan); it's a label with no numerical meaning.

# - Code: Qualitative (nominal)
#    Country ISO code (e.g., AFG); also a label, not a measurable value.

# - Year: Quantitative (discrete)
#    Observation year (e.g., 1990, 1991); numerical integer value.

# - Schizophrenia (%), Bipolar disorder (%), Eating disorders (%),
#   Anxiety disorders (%), Drug use disorders (%), Depression (%), Alcohol use disorders (%)
#    All these columns are:
#   - Quantitative (continuous)
#   - They represent percentages of the population affected by each disorder.
#   - You can perform calculations on them (mean, trends, cross-country comparisons).

#  Summary:
#    Qualitative columns: Entity, Code
#    Quantitative columns: index, Year, and all percentage-based disorder columns
```

```python
#  Data type description for the "Credit Card Approvals" dataset

# - Gender: Qualitative (nominal)
#    Categorical data representing gender (e.g., Male/Female), usually encoded as 0 and 1.

# - Age: Quantitative (continuous)
#    Numerical data representing the applicant's age in years (e.g., 28.46, 37.7).

# - Debt: Quantitative (continuous)
#    Amount of current debt held by the applicant (in thousands); real-valued.

# - Married: Qualitative (nominal)
#    Indicates marital status (e.g., Married/Not Married), encoded as 0 or 1.

# - BankCustomer: Qualitative (nominal)
#    Indicates whether the person is an existing bank customer (yes/no), encoded as binary.

# - YearsEmployed: Quantitative (continuous)
#    Number of years the applicant has been employed; this can include decimal values.

# - PriorDefault: Qualitative (nominal)
#    Indicates whether the applicant has previously defaulted on a loan (0 = No, 1 = Yes).

# - Employed: Qualitative (nominal)
#    Employment status (1 = Employed, 0 = Unemployed).

# - CreditScore: Quantitative (discrete or continuous)
#    Score used to assess creditworthiness. Although it's a number, it could be treated as discrete or continuous depending on granularity.

# - DriversLicense: Qualitative (nominal)
#    Whether the applicant has a driver's license (1 = Yes, 0 = No).

# - ZipCode: Qualitative (nominal)
#    Postal code. Even though it's numeric, it's a categorical variable (no mathematical meaning).

# - Income: Quantitative (continuous)
#    Annual income of the applicant (in dollars), can take a wide range of values.

# - Approved: Qualitative (binary)
#    Target variable: whether the credit card application was approved (1 = Yes, 0 = No).

#  Summary:
# - Quantitative columns: Age, Debt, YearsEmployed, CreditScore, Income
# - Qualitative columns: Gender, Married, BankCustomer, PriorDefault, Employed, DriversLicense, ZipCode, Approved
```

```python
#  Data type description for the Iris dataset

# - Id: Quantitative (discrete)
#    Row identifier (1, 2, 3...); numerical but not meaningful for analysis.

# - SepalLengthCm: Quantitative (continuous)
#    Length of the sepal in centimeters; measurable and allows mathematical operations.

# - SepalWidthCm: Quantitative (continuous)
#    Width of the sepal in centimeters; also a real number.

# - PetalLengthCm: Quantitative (continuous)
#    Length of the petal in centimeters.

# - PetalWidthCm: Quantitative (continuous)
#    Width of the petal in centimeters.

# - Species: Qualitative (nominal)
#    Type of iris flower (e.g., Iris-setosa, Iris-versicolor, Iris-virginica); this is a categorical label.

#  Summary:
# - Quantitative columns: Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
# - Qualitative column: Species
```

```python
# Exercice 5

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Calculate the mean, median, and mode of a quantitative column.

import pandas as pd

file_path = '/content/drive/My Drive/Datasets/Iris.csv'
df = pd.read_csv(file_path)
df.head()


mean_val = df['SepalLengthCm'].mean()       #  Calcule la moyenne (mean) de la colonne 'SepalLengthCm'
median_val = df['SepalLengthCm'].median()   #  Calcule la mdiane (valeur du milieu) de la mme colonne
mode_val = df['SepalLengthCm'].mode()[0]    #  Calcule la ou les valeurs les plus frquentes (mode), on prend la 1re si plusieurs

print("Mean:", mean_val)                    #  Affiche la moyenne dans la console ou le notebook
print("Median:", median_val)                #  Affiche la mdiane
print("Mode:", mode_val)                    #  Affiche le mode
```

```python
#  Cre une nouvelle figure pour le graphique, avec une taille de 8x5 pouces (largeur x hauteur)
plt.figure(figsize=(8,5))

#  Cre un histogramme de la colonne choisie avec :
# - 20 "bacs" (barres) pour la rpartition des valeurs
# - kde=True : ajoute une courbe de densit pour visualiser la tendance globale
sns.histplot(df['SepalLengthCm'], bins=20, kde=True)

#  Ajoute un titre au graphique
plt.title('Distribution of Peta')

#  tiquette pour laxe des X (les valeurs mesures, comme l'ge, le revenu, etc.)
plt.xlabel('Petal Length (cm)')

#  tiquette pour laxe des Y (le nombre d'observations dans chaque intervalle)
plt.ylabel('Nombre dobservations')

#  Affiche le graphique dans le notebook
plt.show()


```
