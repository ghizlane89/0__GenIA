# 1. Understand Classification Types

```python
1. Binary Classification (Classification binaire)
D√©finition : Le mod√®le pr√©dit une seule classe parmi deux possibles.

Exemple : Est-ce qu‚Äôun email est spam ou non-spam ? (classes : 0 ou 1)

--------

2. Multi-class Classification (Classification multi-classes)
D√©finition : Le mod√®le choisit une classe parmi plusieurs (>2).

Exemple : Reconna√Ætre un chiffre manuscrit de 0 √† 9 (10 classes possibles).

------

3. Multi-label Classification
D√©finition : Le mod√®le peut attribuer plusieurs √©tiquettes simultan√©ment √† une m√™me entr√©e.

Exemple : Une image peut contenir un chien üê∂, un chat üê± et une voiture üöó en m√™me temps.
```

# 2. Set Up Your Python Environment and dataset

```python
# 1. Import des biblioth√®ques
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import make_circles

# 2. G√©n√©ration du dataset
samples = 1000
X, y = make_circles(n_samples=samples,
                    noise=0.03,
                    random_state=42)

print("X shape:", X.shape)
print("y shape:", y.shape)

# 3. Visualisation des donn√©es
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.RdYlBu)
plt.title("make_circles dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# 3. Build a Basic Neural Network Model

```python
# 1. Cr√©ation du mod√®le
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid')  # 1 neurone en sortie, pour classe 0 ou 1
])

# 2. Compilation
model.compile(loss='binary_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])

# 3. Entra√Ænement
history = model.fit(X, y, epochs=20)
```

# 4. Improve the Model

```python
# 1. Cr√©ation du mod√®le am√©lior√©
improved_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),  # 1√®re couche cach√©e : 10 neurones
    tf.keras.layers.Dense(10, activation='relu'),  # 2e couche cach√©e : 10 neurones
    tf.keras.layers.Dense(1, activation='sigmoid') # Sortie binaire
])

# 2. Compilation avec Adam
improved_model.compile(loss='binary_crossentropy',
                       optimizer='adam',  # optimiseur adaptatif
                       metrics=['accuracy'])

# 3. Entra√Ænement
improved_history = improved_model.fit(X, y, epochs=100)
```

# 5. Visualize the Decision Boundary

```python
from sklearn.preprocessing import StandardScaler

# Standardisation pour ameliorer le mod√®le

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

improved_history = improved_model.fit(X_scaled, y, epochs=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))

    grid = np.c_[xx.ravel(), yy.ravel()]
    preds = model.predict(grid)
    preds = preds.reshape(xx.shape)

    plt.contourf(xx, yy, preds, levels=[0, 0.5, 1], alpha=0.6, cmap="RdBu")
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", edgecolors='k')
    plt.title("Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

plot_decision_boundary(improved_model, X_scaled, y)
```

# 6. Incorporate Activation Functions

```python
# 1. Cr√©ation du mod√®le avec ReLU (couches cach√©es) et Sigmoid (sortie)
activation_model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu'),   # 1√®re couche cach√©e avec ReLU
    tf.keras.layers.Dense(32, activation='relu'),   # 2e couche cach√©e avec ReLU
    tf.keras.layers.Dense(1, activation='sigmoid')  # Couche de sortie avec Sigmoid
])

# 2. Compilation
activation_model.compile(loss='binary_crossentropy',
                         optimizer='adam',
                         metrics=['accuracy'])

# 3. Entra√Ænement sur les donn√©es standardis√©es
activation_history = activation_model.fit(X_scaled, y, epochs=100)
```

```python
plot_decision_boundary(activation_model, X_scaled, y)

plt.plot(activation_history.history['accuracy'], label='Accuracy')
plt.plot(activation_history.history['loss'], label='Loss')
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.legend()
plt.title("Courbes d'apprentissage")
plt.show()

```

```python
#Commentaire

#Dans cette √©tape, nous avons explor√© l‚Äôimpact des fonctions d‚Äôactivation ReLU et Sigmoid sur les performances d‚Äôun r√©seau de neurones. En int√©grant ReLU dans les couches cach√©es et Sigmoid en sortie, le mod√®le a pu mieux capter la structure non lin√©aire des donn√©es g√©n√©r√©es par make_circles.

#Compar√© aux mod√®les plus simples, cette architecture am√©liore significativement l‚Äôaccuracy et produit une fronti√®re de d√©cision plus fluide et pr√©cise. Cela montre l'importance des fonctions d‚Äôactivation dans la capacit√© d‚Äôun r√©seau √† apprendre des motifs complexes.
```

# 7. Split Data into Training and Testing Sets

```python
# S√©parer les donn√©es

from sklearn.model_selection import train_test_split

# Si tu as d√©j√† standardis√© avec StandardScaler
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Recr√©er et entra√Æner le mod√®le sur X_train

model_split = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_split.compile(loss='binary_crossentropy',
                    optimizer='adam',
                    metrics=['accuracy'])

history_split = model_split.fit(X_train, y_train, epochs=100)
```

```python
#√âvaluer sur le jeu de test

loss_test, accuracy_test = model_split.evaluate(X_test, y_test)
print(f"Test Loss: {loss_test:.4f}")
print(f"Test Accuracy: {accuracy_test:.4f}")
```

```python
# Visualiser la fronti√®re de d√©cision sur l‚Äôensemble complet

plot_decision_boundary(model_split, X_scaled, y)
```

```python
# accuracy sur le test est proche du train
```

```python
# 2. Visualiser la fronti√®re de d√©cision sur le jeu d'entra√Ænement

plot_decision_boundary(model_split, X_train, y_train)
plt.title("Fronti√®re de d√©cision - Donn√©es d'entra√Ænement")
```

# √âtape 9 ‚Äì Apply Regularization Techniques


```python
# Regularization avec Dropout

from tensorflow.keras import layers, regularizers

model_dropout = tf.keras.Sequential([
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),  # 30% des neurones d√©sactiv√©s √† chaque batch
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')
])

model_dropout.compile(loss='binary_crossentropy',
                      optimizer='adam',
                      metrics=['accuracy'])

history_dropout = model_dropout.fit(X_train, y_train, epochs=100)

# √âvaluation
loss_d, acc_d = model_dropout.evaluate(X_test, y_test)
print(f"Test Accuracy avec Dropout : {acc_d:.4f}")
```

```python
# Regularization avec L2 (poids p√©nalis√©s)

model_l2 = tf.keras.Sequential([
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(1, activation='sigmoid')
])

model_l2.compile(loss='binary_crossentropy',
                 optimizer='adam',
                 metrics=['accuracy'])

history_l2 = model_l2.fit(X_train, y_train, epochs=100)

# √âvaluation
loss_l2, acc_l2 = model_l2.evaluate(X_test, y_test)
print(f"Test Accuracy avec L2 : {acc_l2:.4f}")
```

```python
# Courbe de loss pendant l'entra√Ænement

plt.figure(figsize=(12, 5))
plt.plot(history_split.history['loss'], label='Base model (no reg)')
plt.plot(history_dropout.history['loss'], label='Dropout model')
plt.plot(history_l2.history['loss'], label='L2 model')
plt.title('Loss pendant l\'entra√Ænement')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Courbe d‚Äôaccuracy

plt.figure(figsize=(12, 5))
plt.plot(history_split.history['accuracy'], label='Base model (no reg)')
plt.plot(history_dropout.history['accuracy'], label='Dropout model')
plt.plot(history_l2.history['accuracy'], label='L2 model')
plt.title('Accuracy pendant l\'entra√Ænement')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# R√©sum√©:

print(f"Accuracy base : {model_split.evaluate(X_test, y_test, verbose=0)[1]:.4f}")
print(f"Accuracy dropout : {model_dropout.evaluate(X_test, y_test, verbose=0)[1]:.4f}")
print(f"Accuracy L2 : {model_l2.evaluate(X_test, y_test, verbose=0)[1]:.4f}")
```
