```python
## TODO: Calculate the mean and standard deviation of the age for customers who stayed with the bank.

# ğŸ¯ Filtrage des clients qui sont restÃ©s
clients_restes = df[df['Exited'] == 0]

# ğŸ“Š Calcul de la moyenne et de l'Ã©cart-type
mean_age = clients_restes['Age'].mean()
std_age = clients_restes['Age'].std()

# ğŸ–¨ï¸ Affichage des rÃ©sultats
print(f"Ã‚ge moyen des clients restÃ©s : {mean_age:.2f} ans")
print(f"Ã‰cart-type de lâ€™Ã¢ge des clients restÃ©s : {std_age:.2f} ans")
```

### Using Bootstrapping

```python
## TODO: Write a function to perform bootstrap sampling and calculate the statistic of interest.

import numpy as np

def bs_choice(data, func, size):
    """
    RÃ©alise un bootstrap sampling sur les donnÃ©es et applique une fonction statistique Ã  chaque Ã©chantillon.

    ParamÃ¨tres :
    - data : array-like ou pandas Series
    - func : fonction Ã  appliquer (ex: np.mean, np.median, etc.)
    - size : nombre de rÃ©Ã©chantillonnages bootstrap Ã  effectuer

    Retourne :
    - Une liste des statistiques calculÃ©es sur chaque Ã©chantillon bootstrap.
    """
    bs_s = np.empty(size)  # tableau vide pour stocker les rÃ©sultats
    for i in range(size):
        bs_sample = np.random.choice(data, size=len(data), replace=True)  # tirage avec remise
        bs_s[i] = func(bs_sample)  # application de la fonction statistique
    return bs_s



    # Exemple : bootstrap de la moyenne d'Ã¢ge des clients partis
ages_partis = df[df['Exited'] == 1]['Age']
bootstrap_moyennes = bs_choice(ages_partis, np.mean, 1000)

# Calcul d'un intervalle de confiance Ã  95%
ci_low = np.percentile(bootstrap_moyennes, 2.5)
ci_high = np.percentile(bootstrap_moyennes, 97.5)
print(f"Intervalle de confiance Ã  95% pour la moyenne d'Ã¢ge : [{ci_low:.2f}, {ci_high:.2f}]")

```

```python
## TODO: Calculate the difference in means and shift the ages to the overall mean.

# ğŸ¯ Extraction des Ã¢ges
age_restes = df[df['Exited'] == 0]['Age'].values
age_partis = df[df['Exited'] == 1]['Age'].values

# ğŸ“Š Moyennes
mean_restes = np.mean(age_restes)
mean_partis = np.mean(age_partis)
diff_mean = mean_partis - mean_restes
print(f"DiffÃ©rence des moyennes : {diff_mean:.2f}")

# âš–ï¸ Moyenne globale pondÃ©rÃ©e
mean_global = np.mean(np.concatenate([age_restes, age_partis]))

# ğŸ” Recentre les deux groupes autour de la moyenne globale
age_restes_shifted = age_restes - mean_restes + mean_global
age_partis_shifted = age_partis - mean_partis + mean_global

```

```python
## TODO: Perform bootstrap sampling to calculate the standard deviation for both groups and their difference.

import numpy as np

# ğŸ¯ Extraction des Ã¢ges
age_restes = df[df['Exited'] == 0]['Age'].values
age_partis = df[df['Exited'] == 1]['Age'].values

# ğŸ” Fonction de bootstrap sampling pour l'Ã©cart-type
def bs_choice(data, func, size):
    bs_stats = np.empty(size)
    for i in range(size):
        bs_sample = np.random.choice(data, size=len(data), replace=True)
        bs_stats[i] = func(bs_sample)
    return bs_stats

# ğŸ“Š Bootstrap de l'Ã©cart-type pour chaque groupe
bs_std_restes = bs_choice(age_restes, np.std, 1000)
bs_std_partis = bs_choice(age_partis, np.std, 1000)

# ğŸ” Calcul de la distribution de la diffÃ©rence d'Ã©cart-type
bs_diff_std = bs_std_partis - bs_std_restes

# ğŸ“ˆ RÃ©sumÃ©s
print(f"Ã‰cart-type moyen (restÃ©s) : {np.mean(bs_std_restes):.2f}")
print(f"Ã‰cart-type moyen (partis) : {np.mean(bs_std_partis):.2f}")
print(f"DiffÃ©rence moyenne des Ã©cart-types : {np.mean(bs_diff_std):.2f}")

# ğŸ“ Intervalle de confiance Ã  95 %
ci_low = np.percentile(bs_diff_std, 2.5)
ci_high = np.percentile(bs_diff_std, 97.5)
print(f"Intervalle de confiance Ã  95% pour la diffÃ©rence des Ã©carts-types : [{ci_low:.2f}, {ci_high:.2f}]")




```

```python
## TODO: Calculate the p-value by comparing the difference in means to the bootstrap distribution.

# Ã‰tape 1 : DÃ©jÃ  fait prÃ©cÃ©demment
# On suppose que tu as dÃ©jÃ  :
# - age_restes_shifted
# - age_partis_shifted
# - diff_mean (diffÃ©rence rÃ©elle)

# Ã‰tape 2 : Bootstrap de la diffÃ©rence des moyennes sous H0
bs_diff_means = np.empty(1000)
for i in range(1000):
    bs_sample_restes = np.random.choice(age_restes_shifted, size=len(age_restes_shifted), replace=True)
    bs_sample_partis = np.random.choice(age_partis_shifted, size=len(age_partis_shifted), replace=True)
    bs_diff_means[i] = np.mean(bs_sample_partis) - np.mean(bs_sample_restes)

# Ã‰tape 3 : Calcul de la p-value (test bilatÃ©ral)
p_value = np.mean(np.abs(bs_diff_means) >= np.abs(diff_mean))
print(f"DiffÃ©rence rÃ©elle des moyennes : {diff_mean:.2f}")
print(f"p-value bootstrap : {p_value:.4f}")



```

### Conclusion
Do we reject the Null Hypothesis ? Why ?

```python
# âœ… Conclusion:
# Since the p-value (0.011) is less than the threshold of 0.05, we reject the null hypothesis.
# This means the difference in mean age between customers who left and those who stayed is statistically significant.
```

```python
## TODO: Perform a t-test to compare the CreditScore between the two groups (Still with bank and Left the bank).

from scipy.stats import ttest_ind

# ğŸ¯ SÃ©paration des deux groupes selon Exited
score_restes = df[df['Exited'] == 0]['Credit_Score']
score_partis = df[df['Exited'] == 1]['Credit_Score']

# ğŸ§ª Test t (on suppose des variances inÃ©gales => Welch's t-test)
t_stat, p_value = ttest_ind(score_restes, score_partis, equal_var=False)

# ğŸ“Š RÃ©sultats
print(f"Statistique t : {t_stat:.2f}")
print(f"p-value : {p_value:.4f}")

# ğŸ§  InterprÃ©tation
if p_value < 0.05:
    print("âœ… Il y a une diffÃ©rence significative entre les Credit Scores des deux groupes.")
else:
    print("âŒ Aucune diffÃ©rence significative dÃ©tectÃ©e entre les deux groupes.")
```

### Conclusion

### Using Bootstrapping

```python
## TODO: Calculate the difference in means and shift the EstimatedSalary for both groups.

# ğŸ¯ SÃ©paration des deux groupes
salaire_restes = df[df['Exited'] == 0]['Transaction_History'].values
salaire_partis = df[df['Exited'] == 1]['Transaction_History'].values

# ğŸ“Š Moyennes individuelles
mean_restes = salaire_restes.mean()
mean_partis = salaire_partis.mean()

# ğŸ”¢ DiffÃ©rence observÃ©e
diff_mean = mean_partis - mean_restes
print(f"DiffÃ©rence rÃ©elle des moyennes : {diff_mean:.2f}")

# âš–ï¸ Moyenne globale (salaire combinÃ©)
mean_global = np.mean(np.concatenate([salaire_restes, salaire_partis]))

# ğŸ” Centrages sur la moyenne globale
salaire_restes_shifted = salaire_restes - mean_restes + mean_global
salaire_partis_shifted = salaire_partis - mean_partis + mean_global

```

```python
## TODO: Calculate the bootstrap sample means for both groups and their difference.

# ğŸ§ª Bootstrap sampling depuis les distributions recentrÃ©es
n_iterations = 1000
bs_diff_means = np.empty(n_iterations)

for i in range(n_iterations):
    bs_restes = np.random.choice(salaire_restes_shifted, size=len(salaire_restes_shifted), replace=True)
    bs_partis = np.random.choice(salaire_partis_shifted, size=len(salaire_partis_shifted), replace=True)
    bs_diff_means[i] = np.mean(bs_partis) - np.mean(bs_restes)

# ğŸ“ˆ RÃ©sumÃ© des rÃ©sultats
mean_diff_bootstrap = np.mean(bs_diff_means)
ci_low = np.percentile(bs_diff_means, 2.5)
ci_high = np.percentile(bs_diff_means, 97.5)

print(f"DiffÃ©rence moyenne bootstrap : {mean_diff_bootstrap:.2f}")
print(f"Intervalle de confiance Ã  95% : [{ci_low:.2f}, {ci_high:.2f}]")
```

```python
## TODO: Calculate the p-value based on the bootstrap distribution of the difference in means.

# ğŸ“Œ Rappel : diff_mean est la vraie diffÃ©rence observÃ©e
# bs_diff_means est la distribution bootstrap centrÃ©e (sous H0)

# ğŸ¯ Calcul de la p-value (test bilatÃ©ral)
p_value_bootstrap = np.mean(np.abs(bs_diff_means) >= np.abs(diff_mean))

# ğŸ“Š Affichage
print(f"DiffÃ©rence rÃ©elle des moyennes : {diff_mean:.2f}")
print(f"p-value bootstrap : {p_value_bootstrap:.4f}")

# ğŸ§  InterprÃ©tation
if p_value_bootstrap < 0.05:
    print("âœ… On rejette l'hypothÃ¨se nulle : la diffÃ©rence de salaire estimÃ© est significative.")
else:
    print("âŒ On ne rejette pas l'hypothÃ¨se nulle : la diffÃ©rence peut Ãªtre due au hasard.")
```
