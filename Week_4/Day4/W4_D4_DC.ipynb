```python
## TODO: Calculate the mean and standard deviation of the age for customers who stayed with the bank.

# 🎯 Filtrage des clients qui sont restés
clients_restes = df[df['Exited'] == 0]

# 📊 Calcul de la moyenne et de l'écart-type
mean_age = clients_restes['Age'].mean()
std_age = clients_restes['Age'].std()

# 🖨️ Affichage des résultats
print(f"Âge moyen des clients restés : {mean_age:.2f} ans")
print(f"Écart-type de l’âge des clients restés : {std_age:.2f} ans")
```

```python
## TODO: Calculate the mean and standard deviation of the age for customers who left the bank.

# 🎯 Filtrage des clients qui ont quitté la banque
clients_partis = df[df['Exited'] == 1]

# 📊 Calcul de la moyenne et de l'écart-type
mean_age_exited = clients_partis['Age'].mean()
std_age_exited = clients_partis['Age'].std()

# 🖨️ Affichage des résultats
print(f"Âge moyen des clients partis : {mean_age_exited:.2f} ans")
print(f"Écart-type de l’âge des clients partis : {std_age_exited:.2f} ans")
```

```python
## TODO: Perform a t-test to compare the ages of customers who stayed and left the bank.

from scipy.stats import ttest_ind

# 🎯 Séparation des deux groupes
age_restes = df[df['Exited'] == 0]['Age']
age_partis = df[df['Exited'] == 1]['Age']

# 🧪 Test t de Student (variances non supposées égales)
t_stat, p_value = ttest_ind(age_restes, age_partis, equal_var=False)

# 📊 Affichage du résultat
print(f"Statistique t : {t_stat:.2f}")
print(f"p-value : {p_value:.4f}")

# 🧠 Interprétation simple
if p_value < 0.05:
    print("✅ Il y a une différence significative entre les âges des clients partis et restés.")
else:
    print("❌ Aucune différence significative détectée entre les deux groupes.")

```

### Using Bootstrapping

```python
## TODO: Write a function to perform bootstrap sampling and calculate the statistic of interest.

import numpy as np

def bs_choice(data, func, size):
    """
    Réalise un bootstrap sampling sur les données et applique une fonction statistique à chaque échantillon.

    Paramètres :
    - data : array-like ou pandas Series
    - func : fonction à appliquer (ex: np.mean, np.median, etc.)
    - size : nombre de rééchantillonnages bootstrap à effectuer

    Retourne :
    - Une liste des statistiques calculées sur chaque échantillon bootstrap.
    """
    bs_s = np.empty(size)  # tableau vide pour stocker les résultats
    for i in range(size):
        bs_sample = np.random.choice(data, size=len(data), replace=True)  # tirage avec remise
        bs_s[i] = func(bs_sample)  # application de la fonction statistique
    return bs_s



    # Exemple : bootstrap de la moyenne d'âge des clients partis
ages_partis = df[df['Exited'] == 1]['Age']
bootstrap_moyennes = bs_choice(ages_partis, np.mean, 1000)

# Calcul d'un intervalle de confiance à 95%
ci_low = np.percentile(bootstrap_moyennes, 2.5)
ci_high = np.percentile(bootstrap_moyennes, 97.5)
print(f"Intervalle de confiance à 95% pour la moyenne d'âge : [{ci_low:.2f}, {ci_high:.2f}]")

```

```python
## TODO: Calculate the difference in means and shift the ages to the overall mean.

# 🎯 Extraction des âges
age_restes = df[df['Exited'] == 0]['Age'].values
age_partis = df[df['Exited'] == 1]['Age'].values

# 📊 Moyennes
mean_restes = np.mean(age_restes)
mean_partis = np.mean(age_partis)
diff_mean = mean_partis - mean_restes
print(f"Différence des moyennes : {diff_mean:.2f}")

# ⚖️ Moyenne globale pondérée
mean_global = np.mean(np.concatenate([age_restes, age_partis]))

# 🔁 Recentre les deux groupes autour de la moyenne globale
age_restes_shifted = age_restes - mean_restes + mean_global
age_partis_shifted = age_partis - mean_partis + mean_global

```

```python
## TODO: Perform bootstrap sampling to calculate the standard deviation for both groups and their difference.

import numpy as np

# 🎯 Extraction des âges
age_restes = df[df['Exited'] == 0]['Age'].values
age_partis = df[df['Exited'] == 1]['Age'].values

# 🔁 Fonction de bootstrap sampling pour l'écart-type
def bs_choice(data, func, size):
    bs_stats = np.empty(size)
    for i in range(size):
        bs_sample = np.random.choice(data, size=len(data), replace=True)
        bs_stats[i] = func(bs_sample)
    return bs_stats

# 📊 Bootstrap de l'écart-type pour chaque groupe
bs_std_restes = bs_choice(age_restes, np.std, 1000)
bs_std_partis = bs_choice(age_partis, np.std, 1000)

# 🔁 Calcul de la distribution de la différence d'écart-type
bs_diff_std = bs_std_partis - bs_std_restes

# 📈 Résumés
print(f"Écart-type moyen (restés) : {np.mean(bs_std_restes):.2f}")
print(f"Écart-type moyen (partis) : {np.mean(bs_std_partis):.2f}")
print(f"Différence moyenne des écart-types : {np.mean(bs_diff_std):.2f}")

# 📐 Intervalle de confiance à 95 %
ci_low = np.percentile(bs_diff_std, 2.5)
ci_high = np.percentile(bs_diff_std, 97.5)
print(f"Intervalle de confiance à 95% pour la différence des écarts-types : [{ci_low:.2f}, {ci_high:.2f}]")




```

```python
## TODO: Calculate the p-value by comparing the difference in means to the bootstrap distribution.

# Étape 1 : Déjà fait précédemment
# On suppose que tu as déjà :
# - age_restes_shifted
# - age_partis_shifted
# - diff_mean (différence réelle)

# Étape 2 : Bootstrap de la différence des moyennes sous H0
bs_diff_means = np.empty(1000)
for i in range(1000):
    bs_sample_restes = np.random.choice(age_restes_shifted, size=len(age_restes_shifted), replace=True)
    bs_sample_partis = np.random.choice(age_partis_shifted, size=len(age_partis_shifted), replace=True)
    bs_diff_means[i] = np.mean(bs_sample_partis) - np.mean(bs_sample_restes)

# Étape 3 : Calcul de la p-value (test bilatéral)
p_value = np.mean(np.abs(bs_diff_means) >= np.abs(diff_mean))
print(f"Différence réelle des moyennes : {diff_mean:.2f}")
print(f"p-value bootstrap : {p_value:.4f}")



```

### Conclusion
Do we reject the Null Hypothesis ? Why ?

```python
# ✅ Conclusion:
# Since the p-value (0.011) is less than the threshold of 0.05, we reject the null hypothesis.
# This means the difference in mean age between customers who left and those who stayed is statistically significant.
```

```python
## TODO: Perform a t-test to compare the CreditScore between the two groups (Still with bank and Left the bank).

from scipy.stats import ttest_ind

# 🎯 Séparation des deux groupes selon Exited
score_restes = df[df['Exited'] == 0]['Credit_Score']
score_partis = df[df['Exited'] == 1]['Credit_Score']

# 🧪 Test t (on suppose des variances inégales => Welch's t-test)
t_stat, p_value = ttest_ind(score_restes, score_partis, equal_var=False)

# 📊 Résultats
print(f"Statistique t : {t_stat:.2f}")
print(f"p-value : {p_value:.4f}")

# 🧠 Interprétation
if p_value < 0.05:
    print("✅ Il y a une différence significative entre les Credit Scores des deux groupes.")
else:
    print("❌ Aucune différence significative détectée entre les deux groupes.")
```

### Conclusion
Do we reject the Null Hypothesis ? Why ?

```python
# ✅ Conclusion :
# Le test t indique une p-value de 0.0014, bien inférieure au seuil de 0.05.
# On rejette donc l’hypothèse nulle : le Credit Score a une influence significative sur le churn client.
```

### Using Bootstrapping

```python
## TODO: Calculate the difference in means and shift the EstimatedSalary for both groups.

# 🎯 Séparation des deux groupes
salaire_restes = df[df['Exited'] == 0]['Transaction_History'].values
salaire_partis = df[df['Exited'] == 1]['Transaction_History'].values

# 📊 Moyennes individuelles
mean_restes = salaire_restes.mean()
mean_partis = salaire_partis.mean()

# 🔢 Différence observée
diff_mean = mean_partis - mean_restes
print(f"Différence réelle des moyennes : {diff_mean:.2f}")

# ⚖️ Moyenne globale (salaire combiné)
mean_global = np.mean(np.concatenate([salaire_restes, salaire_partis]))

# 🔁 Centrages sur la moyenne globale
salaire_restes_shifted = salaire_restes - mean_restes + mean_global
salaire_partis_shifted = salaire_partis - mean_partis + mean_global

```

```python
## TODO: Calculate the bootstrap sample means for both groups and their difference.

# 🧪 Bootstrap sampling depuis les distributions recentrées
n_iterations = 1000
bs_diff_means = np.empty(n_iterations)

for i in range(n_iterations):
    bs_restes = np.random.choice(salaire_restes_shifted, size=len(salaire_restes_shifted), replace=True)
    bs_partis = np.random.choice(salaire_partis_shifted, size=len(salaire_partis_shifted), replace=True)
    bs_diff_means[i] = np.mean(bs_partis) - np.mean(bs_restes)

# 📈 Résumé des résultats
mean_diff_bootstrap = np.mean(bs_diff_means)
ci_low = np.percentile(bs_diff_means, 2.5)
ci_high = np.percentile(bs_diff_means, 97.5)

print(f"Différence moyenne bootstrap : {mean_diff_bootstrap:.2f}")
print(f"Intervalle de confiance à 95% : [{ci_low:.2f}, {ci_high:.2f}]")
```

```python
## TODO: Calculate the p-value based on the bootstrap distribution of the difference in means.

# 📌 Rappel : diff_mean est la vraie différence observée
# bs_diff_means est la distribution bootstrap centrée (sous H0)

# 🎯 Calcul de la p-value (test bilatéral)
p_value_bootstrap = np.mean(np.abs(bs_diff_means) >= np.abs(diff_mean))

# 📊 Affichage
print(f"Différence réelle des moyennes : {diff_mean:.2f}")
print(f"p-value bootstrap : {p_value_bootstrap:.4f}")

# 🧠 Interprétation
if p_value_bootstrap < 0.05:
    print("✅ On rejette l'hypothèse nulle : la différence de salaire estimé est significative.")
else:
    print("❌ On ne rejette pas l'hypothèse nulle : la différence peut être due au hasard.")
```
